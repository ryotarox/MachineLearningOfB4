```python
%matplotlib inline
import numpy as np
np.random.seed(555)
import matplotlib.pyplot as plt
import pandas as pd

from statsmodels.graphics import tsaplots
from statsmodels.tsa import stattools
from statsmodels.tsa import ar_model
```

```python
import statsmodels.api as sm # version 0.8.0以上
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
status = pd.read_csv('../../../status.csv')
```

```python
#statusのyear, month, dayを結合してdatetime型に
status['date'] = status['year'].astype(str) + '/' + status['month'].astype(str).str.zfill(2).astype(str) + '/' + status['day'].astype(str).str.zfill(2).astype(str) + '/' + status['hour'].astype(str).str.zfill(2).astype(str)
status['date'] = pd.to_datetime(status['date'], format='%Y/%m/%d/%H', infer_datetime_format=True)

#曜日を追加するための関数を定義
def get_weekday_jp(dt):
    w_list = ['月曜日', '火曜日', '水曜日', '木曜日', '金曜日', '土曜日', '日曜日']
    return(w_list[dt.weekday()])

#dateから曜日情報を取得
status["weekday"] = status["date"].apply(get_weekday_jp)

main_df = status[['date','hour', 'station_id', 'bikes_available', 'weekday', 'predict']]

#カテゴリ変数をダミー変数化
main_df = pd.get_dummies(main_df)

train = main_df[main_df['date'] >= '2013-10-01']
train = train[train['date'] < '2013-11-01']
train = train[train['station_id'] == 0] 

test = main_df[main_df['date'] >= '2013-11-01']
test = test[test['date'] <= '2013-11-08 00:00:00']
test = test[test['station_id'] == 0] 

#predictは特徴量として必要ないため、削除
train = train.drop("predict",axis=1)
train_df = train 
train
```

```python
status_v1 = train.set_index('date')
```

```python
#predictは特徴量として必要ないため、削除
status_v1 = status_v1.drop("hour",axis=1)
status_v1 = status_v1.drop("station_id",axis=1)
status_v1
```

```python
status_exog =  status_v1.drop("bikes_available",axis=1)
status_exog
```

```python
p, d, q, sp, sd, sq = 4, 1, 5, 1, 1, 1
seasonal = 24

# SARIMAモデルの当てはめいろいろ
result_sarima_1 = SARIMAX(
    endog = status_v1["bikes_available"], 
    order=(p,d,q), 
    seasonal_order=(sp,sd,sq,seasonal), 
    enforce_stationarity = False, 
    enforce_invertibility = False
).fit()
# 結果確認
print(result_sarima_1.summary())
```

```python
# 将来予測
plt.figure(figsize=(16,8))
plt.plot(status_v1["bikes_available"], label="observation")
plt.plot(result_sarima_1.fittedvalues, color='red')
predict_sarima_1 = result_sarima_1.predict('2013-11-01', '2013-11-08')
plt.plot(predict_sarima_1) # 予測データをプロット
```

```python
p, d, q, sp, sd, sq = 4, 1, 5, 1, 1, 1
seasonal = 24

# SARIMAモデルの当てはめいろいろ
result_sarima_2 = SARIMAX(
    endog = status_v1["bikes_available"], 
    exog =status_exog,
    order=(p,d,q), 
    seasonal_order=(sp,sd,sq,seasonal), 
    enforce_stationarity = False, 
    enforce_invertibility = False
).fit()
# 結果確認
print(result_sarima_2.summary())
```

```python
#predictは特徴量として必要ないため、削除
exog_forecast = test.drop("predict",axis=1)
exog_forecast = exog_forecast.drop("hour",axis=1)
exog_forecast = exog_forecast.drop("station_id",axis=1)
exog_forecast = exog_forecast.drop("bikes_available",axis=1)
exog_forecast = exog_forecast.set_index('date')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
train_basic = status_v1['bikes_available']
```

```python
decomposition_train = seasonal_decompose(train_basic)
trend_train = decomposition_train.trend
seasonal_train = decomposition_train.seasonal
residual_train = decomposition_train.resid

plt.figure(figsize=(16,12))
plt.subplot(411)
plt.plot(train_basic, label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend_train, label='Trend')
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal_train,label='Seasonality')
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual_train, label='Residuals')
plt.legend(loc='best')
plt.tight_layout()
```
